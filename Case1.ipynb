```python
import pandas as pd
from google.colab import drive

drive.mount('/content/drive')

file_path = '/content/drive/MyDrive/creditcard.csv'

df = pd.read_csv(file_path)

print(df.head())
```

```python
print(df.shape)
```

```python
df.head()
```

```python
X = df.drop(columns=['Class'])
X.head()
y = df.Class
print(X)
print(y)
```

```python
class_counts = df['Class'].value_counts()
class_counts
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from imblearn.under_sampling import RandomUnderSampler
rus=RandomUnderSampler()
X_train,y_train=rus.fit_resample(X_train,y_train)
y_train.value_counts()
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

print("Accuracy: ", accuracy)
print("Precision: ", precision)
print("Recall: ", recall)
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report

# K-Nearest Neighbors Classifier
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)
knn_report = classification_report(y_test, y_pred_knn)
print('K-NN Classifier Performance:')
print(knn_report)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

# Logistic Regression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred_logreg = logreg.predict(X_test)
logreg_report = classification_report(y_test, y_pred_logreg)
print('Logistic Regression Performance:')
print(logreg_report)

# Naive Bayes
nb = GaussianNB()
nb.fit(X_train, y_train)
y_pred_nb = nb.predict(X_test)
nb_report = classification_report(y_test, y_pred_nb)
print('Naive Bayes Performance:')
print(nb_report)

# Support Vector Machine
svm = SVC()
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_test)
svm_report = classification_report(y_test, y_pred_svm)
print('SVM Performance:')
print(svm_report)
```

```python
from sklearn.metrics import accuracy_score
import pandas as pd

# Compute accuracies
accuracy_logreg = accuracy_score(y_test, y_pred_logreg)
accuracy_nb = accuracy_score(y_test, y_pred_nb)
accuracy_svm = accuracy_score(y_test, y_pred_svm)
accuracy_knn = accuracy_score(y_test, y_pred_knn)

# Create a DataFrame to display the accuracies
accuracy_df = pd.DataFrame({
    'Model': ['K-NN', 'Logistic Regression', 'Naive Bayes', 'Support Vector Machine'],
    'Accuracy': [accuracy_knn, accuracy_logreg, accuracy_nb, accuracy_svm]
})

# Print the table
print("Model Comparison - Accuracy:")
accuracy_df
```
